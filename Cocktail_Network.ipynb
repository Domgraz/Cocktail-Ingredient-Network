```python
import requests 
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

I asked ChatGPT to give me a 50 popular cocktails which I will get the individual recipes for

```python
cocktail_list = ["Margarita", "Martini", "Mojito", "Cosmopolitan", "Old Fashioned", "Manhattan", "Pina Colada", "Mai Tai", "Daiquiri", "Moscow Mule", "Bloody Mary", "Negroni", "Whiskey Sour", "Pi√±a Colada", "Tequila Sunrise", "Sidecar", "White Russian", "Long Island Iced Tea", "Sex on the Beach", "Blue Lagoon", "Singapore Sling", "Sazerac", "Tom Collins", "White Lady", "Zombie", "Planter's Punch", "Grasshopper", "Caipirinha", "Bellini", "B-52", "Alabama Slammer", "Dark 'n' Stormy", "Hurricane", "Irish Coffee", "Tequila Mockingbird", "French 75", "Boulevardier", "Paloma", "Bee's Knees", "Vesper", "Aviation", "Painkiller", "Corpse Reviver", "Sazerac", "Rob Roy", "Caipiroska", "Bramble", "Rusty Nail", "Blue Hawaii", "Rusty Nail",'Rum and Coke']
```

Uses the cocktaildb API to get the data on each of the cocktails above, and creates the initial dataframe

```python
cocktail_df = pd.DataFrame(columns=["Cocktail Name", "Ingredients"])

# Loop through the list of cocktail names and retrieve their data
for cocktail_name in cocktail_list:
    # Make an API request to get cocktail details
    url = f"https://www.thecocktaildb.com/api/json/v1/1/search.php?s={cocktail_name}"
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        cocktails = data["drinks"]

        if cocktails:
            cocktail = cocktails[0]
            ingredients = []

            # Loop through the ingredients
            for i in range(1, 16):
                ingredient = cocktail.get(f"strIngredient{i}")
                measure = cocktail.get(f"strMeasure{i}")
                
                if ingredient:
                    ingredients.append(f"{measure} {ingredient}")

            # Add data to the DataFrame
            cocktail_df = cocktail_df.append({"Cocktail Name": cocktail_name, "Ingredients": ", ".join(ingredients)}, ignore_index=True)
        else:
            print(f"Cocktail '{cocktail_name}' not found.")
    else:
        print(f"Failed to retrieve data for '{cocktail_name}'. Status code:", response.status_code)
```

```python
cocktail_df.head()
```

```python
ingredients_split = cocktail_df['Ingredients'].str.split(', ', expand=True)

# Concatenate the 'Cocktail Name' and the split ingredients
result_df = pd.concat([cocktail_df['Cocktail Name'], ingredients_split], axis=1)

# Rename the columns for clarity
result_df.columns = ['Cocktail Name'] + [f'Ingredient {i+1}' for i in range(ingredients_split.shape[1])]
```

```python
result_df.head(50)
```

```python
result_df = result_df.replace(to_replace=[r'\d',r'/', r' oz', r'tblsp',r'-',r'. cL',r'tsp',r'cubes',r'cube',r'dashes',r'dash',r'cl',r'parts',r'part',r'measures',r'measure',r'\.',r'jiggers',r'juice',r'\bof\b(?! [A-Za-z])',r'Juice',r'cup',r'None',r'\bor\b(?! [A-Za-z])',r'Twist',r'twist',r'drops',r'drop'], value='', regex=True)
```

```python
result_df.head(50)
```

```python
G = nx.Graph()
for col in result_df.columns[1:]:
    edges = [(row['Cocktail Name'], row[col]) for _, row in result_df.iterrows() if not pd.isna(row[col])]
    G.add_edges_from(edges)

degrees = dict(G.degree())

# Sort nodes based on their degree in descending order
sorted_nodes = sorted(degrees, key=lambda x: degrees[x], reverse=True)

# Set up the layout for node positioning
pos = nx.spring_layout(G,scale = 2.5,k = .168)
node_colors = {node: 'skyblue' if node in cocktail_list else 'lightgreen' for node in G.nodes()}
node_sizes = [1700 if node in cocktail_list else 700 for node in G.nodes()]
# Increase the figure size
plt.figure(figsize=(50, 50))


# Draw nodes and edges
nx.draw(G, pos, with_labels=True, node_color=[node_colors[node] for node in G.nodes()], node_size=node_sizes, font_size=25, width=2)

# Draw labels with automatic placement
labels = nx.draw_networkx_labels(G, pos, font_size=25, font_color='black')
#nx.draw(G, pos,with_labels=True, node_color=[node_colors[node] for node in G.nodes()], node_size = 700, font_size=25,width = 2)
plt.title("Cocktails and their Ingredients",fontsize = 100)
plt.show()
```

Get the top 3 nodes in the graph accoring to the highest degree centrality

```python
degree_cent = nx.degree_centrality(G)
#sort to get the highest
sort_node = sorted(degree_cent, key = degree_cent.get,reverse=True)
top_3 = sort_node[:3]
print(top_3)
```
